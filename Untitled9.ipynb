```python
import pandas as pd
from sklearn import datasets

iris = datasets.load_iris()
X = iris.data
y = iris.target
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.DataFrame(X, columns=iris.feature_names)
df['species'] = iris.target

sns.pairplot(df, hue='species')
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_standardized = scaler.fit_transform(X)
```

```python
import numpy as np

covariance_matrix = np.cov(X_standardized.T)
```

```python

eigenvalues, eigenvectors = np.linalg.eig(covariance_matrix)
```

```python

sorted_indices = np.argsort(eigenvalues)[::-1]
sorted_eigenvalues = eigenvalues[sorted_indices]
sorted_eigenvectors = eigenvectors[:, sorted_indices]

X_pca = X_standardized.dot(sorted_eigenvectors[:, :2])
```

```python

plt.figure(figsize=(8, 6))
colors = ['navy', 'turquoise', 'darkorange']
for color, i, target_name in zip(colors, [0, 1, 2], iris.target_names):
    plt.scatter(X_pca[y == i, 0], X_pca[y == i, 1], color=color, alpha=0.8, label=target_name)
plt.title('PCA of IRIS dataset')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(loc='best', shadow=False, scatterpoints=1)
plt.show()
```

```python
import matplotlib.pyplot as plt
from skimage import io, color

image_url = '/content/sentosaimg.jpg'
image = io.imread(image_url)
gray_image = color.rgb2gray(image)
plt.imshow(gray_image, cmap='gray')
plt.title('Original Image')
plt.axis('off')
plt.show()
```

```python

original_shape = gray_image.shape
flattened_image = gray_image.reshape(original_shape[0], -1)  # Flattening the image
```

```python
import numpy as np

U, S, Vt = np.linalg.svd(flattened_image, full_matrices=False)
```

```python
def reconstruct_image_svd(U, S, Vt, num_components):

    U_reduced = U[:, :num_components]
    S_reduced = np.diag(S[:num_components])
    Vt_reduced = Vt[:num_components, :]


    return np.dot(U_reduced, np.dot(S_reduced, Vt_reduced))

components_list = [5, 20, 50, 100]
reconstructed_images_svd = []

for n in components_list:
    reconstructed_images_svd.append(reconstruct_image_svd(U, S, Vt, n))


reconstructed_images_svd.insert(0, gray_image)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 8))

titles = ['Original Image'] + [f'Reconstructed Image with {n} Singular Values' for n in components_list]

num_rows = 2
num_cols = (len(reconstructed_images_svd) + 1) // 2

for i in range(len(reconstructed_images_svd)):

    plt.subplot(num_rows, num_cols, i + 1)
    plt.imshow(reconstructed_images_svd[i], cmap='gray')
    plt.title(titles[i])
    plt.axis('off')

plt.tight_layout()
plt.show()
```
